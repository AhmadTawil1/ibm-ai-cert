## üß¨ Biological Neurons: The Inspiration for Deep Learning

All algorithms used in deep learning are **inspired by how neurons in the human brain process data**. The key structure of a biological neuron includes:

- **Soma**: The main body that contains the **nucleus** (processing center).
- **Dendrites**: Branching arms that receive electrical signals (input).
- **Axon**: A long extension that sends signals (output).
- **Synapses**: Terminal buttons at the end of the axon that connect to other neurons.

> üñº The earliest neuron drawing was made by **Santiago Ram√≥n y Cajal** in 1899, based on a pigeon‚Äôs brain under a microscope. He is known as the *father of modern neuroscience*.

---

## üîÑ How a Biological Neuron Works

1. **Dendrites** receive incoming signals (electrical impulses) from other neurons.
2. Signals are passed to the **soma** where they are **processed**.
3. The **axon** carries the processed signal away from the soma.
4. **Synapses** transmit the signal to **other neurons**.

> This flow of information is directional:  
> **Input (Dendrites) ‚Üí Processing (Soma) ‚Üí Output (Axon ‚Üí Synapse)**

## ü§ñ Artificial Neurons: The Core of Deep Learning

Artificial neurons **mimic biological neurons** in structure and function:

- **Inputs (x‚ÇÅ, x‚ÇÇ, ..., x‚Çô)** are like dendrites.
- **Weighted sum + bias** represents processing in the soma.
- **Activation function** mimics the firing decision.
- **Output** connects to other neurons (next layer).

### üß† Artificial Neuron Formula

The output of an artificial neuron is computed in two steps:

1. **Weighted sum of inputs**:

$$
z = \sum_{i=1}^{n} w_i x_i + b
$$

2. **Activation function applied**:

$$
a = \sigma(z)
$$

Where:  
- \( x_i \) = input features  
- \( w_i \) = weights  
- \( b \) = bias  
- \( \sigma \) = activation function (e.g., ReLU, Sigmoid)  
- \( a \) = output of the neuron

## üìö Key Takeaways from the Video

- **Neural networks** in deep learning are modeled after the brain's network of neurons.
- The **soma** processes input signals and determines whether to "fire."
- **Learning** happens by **strengthening connections** (weights) based on useful patterns.
- Artificial neurons learn by **adjusting weights** during training to improve prediction accuracy.

## üß† Fun Fact

> ‚ÄúLearning‚Äù in the brain and in artificial networks both happen by **reinforcing useful connections**. This is the basis for **backpropagation** in artificial neural networks.
